```python
del df['product_link']
del df['img_link']
del df['meta_description']
del df['meta_keyword']
```

#### 1)Calculate average rating of products (non-zero)

```python
rat=0
num=0

for i in df['rating']:
    if i!=0:
        rat+=i
        num+=1
```

```python
print("Average is:",round(rat/num))
```
